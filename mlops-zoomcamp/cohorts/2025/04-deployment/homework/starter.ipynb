```python
import pickle
import pandas as pd
```

```python
with open('model.bin', 'rb') as f_in:
    dv, model = pickle.load(f_in)
```

```python
categorical = ['PULocationID', 'DOLocationID']

def read_data(filename):
    df = pd.read_parquet(filename)
    
    df['duration'] = df.tpep_dropoff_datetime - df.tpep_pickup_datetime
    df['duration'] = df.duration.dt.total_seconds() / 60

    df = df[(df.duration >= 1) & (df.duration <= 60)].copy()

    df[categorical] = df[categorical].fillna(-1).astype('int').astype('str')
    
    return df
```

```python
df = read_data('https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2023-04.parquet')
df
```

```python
dicts = df[categorical].to_dict(orient='records')
X_val = dv.transform(dicts)
y_pred = model.predict(X_val)
```

```python
# Set up homework submission
year = 2023
month = 3
```

```python
# Question 1
duration_std_2023_03 = y_pred.std()
print(duration_std_2023_03)
```

```python
# Question 2
df['ride_id'] = f'{year:04d}/{month:02d}_' + df.index.astype('str')
df
```

```python
df_result = pd.DataFrame()
df_result['ride_id'] = df['ride_id']
df_result['predicted_duration'] = y_pred
```

```python
output_file = f'yellow_tripdata_{year:04d}-{month:02d}_predictions.parquet'
```

```python
df_result.to_parquet(
    output_file,
    engine='pyarrow',
    compression=None,
    index=False
)
```
